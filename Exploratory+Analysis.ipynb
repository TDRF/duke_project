```python
count=pd.DataFrame(person['year_of_birth'].value_counts())
row=pd.DataFrame(count._stat_axis.values.tolist())
count=np.array(count)
row=np.array(row)
count_=np.hstack((row,count))
count_=count_[np.lexsort(count_[:,::-1].T)]
count=pd.DataFrame(count_)
count.set_index([0],inplace=True)
count.plot(kind='bar',figsize=(20,5),title='Distribution of birth year')
```

```python
##visit occurrence

start=visit_occurrence['visit_start_date']
end=visit_occurrence['visit_end_date']
days=np.zeros(shape=(1,start.shape[0]))
for i in range(start.shape[0]):
    date1=time.strptime(start[i],"%Y-%m-%d")
    date2=time.strptime(end[i],"%Y-%m-%d")
    date1=datetime.datetime(date1[0],date1[1],date1[2])
    date2=datetime.datetime(date2[0],date2[1],date2[2])
    days[0,i]=(date2-date1).days
pd.Series(days[0,]).value_counts().plot(kind='bar',title='visit_occurrence in days')
```
