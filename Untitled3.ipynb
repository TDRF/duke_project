```python
device_exposure.shape
```

```python
device_exposure=pd.merge(device_exposure,goldstandard, on='person_id', how='outer')
device_exposure['concept']=device_exposure['device_concept_id'].apply(lambda x: 'Ventilator' if x==45768197 else '')
device_exposure['type']=device_exposure['device_type_concept_id'].apply(lambda x: 'EHR Detail' if x==44818707 else '')
device_exposure.head()
```

```python
device_exposure['quantity'].value_counts()
```

```python
device_exposure['concept'].value_counts()
```

```python
device_exposure['type'].value_counts()
```

```python
start=device_exposure['device_exposure_start_date']
end=device_exposure['device_exposure_end_date']
days=np.zeros(shape=(1,start.shape[0]))
for i in range(start.shape[0]):
    date1=time.strptime(start[i],'%Y-%m-%d')
    date2=time.strptime(end[i],'%Y-%m-%d')
    date1=datetime.datetime(date1[0],date1[1],date1[2])
    date2=datetime.datetime(date2[0],date2[1],date2[2])
    days[0,i]=(date2-date1).days

device_exposure['time']=days[0,]

plt.title('device_exposure time in days')
plt.hist(days[0,],bins=50)
```

```python
i=1
plt.figure(figsize=(12,9))
plt.title('device exposure time')
for k in device_exposure['time'].value_counts().index:
    plt.subplot(3,4,i)
    plt.title(k)
    plt.pie(x=device_exposure[device_exposure['time']==k]['status'].value_counts(),labels=device_exposure[device_exposure['time']==k]['status'].value_counts().index,autopct='%.0f%%')
    i=i+1
```
